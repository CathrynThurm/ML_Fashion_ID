```python
#defining axis in a plot grid
W_grid = 15
L_grid = 15
```

```python
#creates subplots on grid
fig, axes = plt.subplots(L_grid, W_grid, figsize = (20, 20))
axes = axes.ravel()
len_training = len(training) #gets length of training set
#select a random number from 0 to len_training
for i in np.arange(0, L_grid * W_grid):
    index = np.random.randint(0, len_training)
    #read and display random image with the selected index
    axes[i].imshow(training[index, 1:].reshape(28, 28))
    axes[i].set_title(training[index, 0], fontsize = 8)
    axes[i].axis('off')
plt.subplots_adjust(hspace = .5)
```

```python
x_train = training[:, 1:]/255
y_train = training[:, 0]
```

```python
x_test = testing[:, 1:]/255
y_test = testing[:, 0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_validate, y_train, y_validate = train_test_split(x_train, y_train, test_size=0.2, random_state=1750)
```

```python
x_train = x_train.reshape(x_train.shape[0], *(28, 28, 1))
x_test = x_test.reshape(x_test.shape[0], *(28, 28, 1))
x_validate = x_validate.reshape(x_validate.shape[0], *(28, 28, 1))
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import TensorBoard
```

```python
cnn_model = Sequential()
```

```python
cnn_model.add(Conv2D(32, 3, 3, input_shape = (28, 28, 1), activation = 'relu' ))
```

```python
cnn_model.add(MaxPooling2D(pool_size = (2, 2)))
```

```python
cnn_model.add(Flatten())
```

```python
cnn_model.add(Dense(64, activation = 'relu'))
```

```python
cnn_model.add(Dense(20, activation = 'sigmoid'))
```

```python
cnn_model.compile(loss = 'sparse_categorical_crossentropy', optimizer = Adam(lr =0.001), metrics = ['accuracy'])
```

```python
epochs = 75
```

```python
cnn_model.fit(x_train, y_train, batch_size = 512, epochs = epochs, verbose = 1, validation_data = (x_validate, y_validate))
```

```python
evaluation = cnn_model.evaluate(x_test, y_test)
```

```python
predicted_classes = cnn_model.predict_classes(x_test)
```

```python
predicted_classes
```

```python
L= 5
W = 5
fog, axes = plt.subplots(L, W, figsize = (12, 12))
axes = axes.ravel()
for i in np.arange(0, L*W):
    axes[i].imshow(x_test[i].reshape(28, 28))
    axes[i].set_title('Predicted class: {:0.1f}\n True class: {:0.1f}'.format(predicted_classes[i], y_test[i]))
    axes[i].axis("off")
plt.subplots_adjust(hspace = .6)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, predicted_classes)
plt.figure(figsize = (14,10))
sns.heatmap(cm, annot = True)
```
